```python
import torch
from transformers import AutoModelForCausalLM,GPT2LMHeadModel
```

```python
# load pretrained GPT-2 model and tokenizer
gpt2 = AutoModelForCausalLM.from_pretrained('gpt2')
```

# IMPORTANT NOTE:
### The methods 'save_pretrained' and 'from_pretrained' are coded into the HuggingFace PreTrainedModel API.
### The code below won't work for custom models or models built in other ecosystems.

# Save the original model, then change and save it again

```python
gpt2.save_pretrained('gptOriginal_HF')
# explore the folder...
```

```python
gpt2.config
```

```python
gpt2.transformer.wte.weight
```

```python
# replace embeddings with 1's
gpt2.transformer.wte.weight.data = torch.ones_like(gpt2.transformer.wte.weight)
gpt2.transformer.wte.weight
```

```python
gpt2.save_pretrained('gptWithOnes_HF')
```

# Clear the model from the workspace

```python
del gpt2
```

# Import the two models from disk

```python
origmodel = GPT2LMHeadModel.from_pretrained('gptOriginal_HF')
onesmodel = GPT2LMHeadModel.from_pretrained('gptWithOnes_HF')
```

```python
origmodel.transformer.wte.weight
```

```python
onesmodel.transformer.wte.weight
```

# Saving and loading the model state (non-HF models)

```python
# pytorch's method for saving the model's state_dict
torch.save(origmodel.state_dict(),'gptOriginalStateDict.pt')
```

```python
# load it back in -- replacing the model with all 1's
onesmodel.load_state_dict(torch.load('gptOriginalStateDict.pt'))
onesmodel.transformer.wte.weight
```
