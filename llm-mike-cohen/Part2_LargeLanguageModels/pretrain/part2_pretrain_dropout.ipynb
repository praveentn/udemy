```
# import libraries
import torch
import torch.nn as nn
import torch.nn.functional as F
```

# Using dropout

```
# define a dropout instance and make some data
prob = .2

dropout = nn.Dropout(p=prob)
x = torch.ones(100)
print('Initial data vector:\n  ',x,'\n')

# let's see what dropout returns
y = dropout(x)
print('"Raw" post-dropout vector:\n  ',y,'\n')

# normalized
print('Normed post-dropout vector:\n  ',y*(1-prob),'\n')

# how many were dropped?
print(f'{100*torch.sum(y==0)/len(x):.1f}% dropped (expected {prob*100:.1f}%)')
```

# Using the function

```
F.dropout(x,p=prob)
```

# Toggling dropout

```
# dropout is turned off when evaluating the model
dropout.eval()
y = dropout(x)
print(y)
```

```
# F.dropout() is not deactivated in eval mode:

dropout.eval()
y = F.dropout(x)
print(y)
```

```
# but you can manually switch it off
# dropout.eval() # no effect :o
y = F.dropout(x,training=False)
print(y)
```

```
# the model needs to be reset after toggling into eval mode

dropout.train()
y = dropout(x)
print('train() mode on:\n',y,'\n') # with dropout


dropout.eval()
y = dropout(x)
print('eval() mode on:\n',y,'\n') # without dropout


dropout.train()
y = dropout(x)
print('eval() remains on:\n',y) # still w/o dropout ;)
```
