```python
from datasets import load_dataset
import textwrap
import numpy as np
import requests
```

# WEB-GLM Q&A

```python
# https://huggingface.co/datasets/THUDM/webglm-qa
dataset = load_dataset("THUDM/webglm-qa")
dataset
```

```python
dataset['test'][10]
```

```python
# pick a random item from the training set
idx = np.random.randint(dataset['train'].num_rows)
print(f'Element {idx}\n')

# print its data
for quest,txt in dataset['train'][idx].items():

  # print the question
  print(f':: {quest} ::')

  # print the answer (its one string)
  if isinstance(txt,str):
    print(f'  {textwrap.fill(txt,88)}\n')

  # print the references (comes in a list)
  elif isinstance(txt,list):
    for i in range(len(txt)):
      print(f'  r{i}: {textwrap.fill(txt[i],88)}\n')
```

# Anthropic's Helpful and Harmless alignment dataset

```python
# https://huggingface.co/datasets/Unified-Language-Model-Alignment/Anthropic_HH_Golden
dataset = load_dataset('Unified-Language-Model-Alignment/Anthropic_HH_Golden')
dataset
```

```python
# pick a random item from the training set
idx = np.random.randint(dataset['train'].num_rows)
print(f'Element {idx}\n')

# print its data
print('*'*20,'  CHOSEN ANSWER  ','*'*20)
print(textwrap.fill(dataset['train'][idx]['chosen'],99))
print('\n\n','*'*20,'  REJECTED ANSWER  ','*'*20)
print(textwrap.fill(dataset['train'][idx]['rejected'],99))
```
