```python
# identify the target token
target_token_him = tokenizer.encode(' him')
target_token_her = tokenizer.encode(' her')
target_token_round = tokenizer.encode(' round')

print(f'The target token indices are {target_token_him}, {target_token_her}, and {target_token_round}\n')

# need to specify a padding token
tokenizer.pad_token = tokenizer.eos_token

# tokenize
tokens = tokenizer(sentences,padding=True,return_tensors='pt')

# example
print(tokens['input_ids'][10])
print(tokens['attention_mask'][10])
```

# Get the activations and hidden states

```python
with torch.no_grad():
  outputs = model(**tokens,output_hidden_states=True)
```

```python
nLayers = model.config.n_layer+1
```

```python
outputs.hidden_states[3].shape
```

# Get the trajectories in embedding space

```python
# initializations
all_acts = np.zeros((nLayers*len(sentences),model.config.n_embd))
lookuptable = np.zeros((all_acts.shape[0],3),dtype=int)
rowi = 0


# loop over sentences
for senti in range(len(sentences)):

  # find the index of either of the target tokens
  sentenceTokens = tokens['input_ids'][senti].numpy()
  targBool = (sentenceTokens==target_token_him) | (sentenceTokens==target_token_her) | (sentenceTokens==target_token_round)
  targidx = np.where(targBool)[0]

  for layeri in range(nLayers):

    # get the activation vector for target token in this layer in this sentence
    all_acts[rowi,:] = outputs.hidden_states[layeri][senti,targidx,:].numpy().squeeze()

    # identifiers
    lookuptable[rowi,0] = senti              # which sentence
    lookuptable[rowi,1] = layeri             # which layer
    lookuptable[rowi,2] = sentenceIDs[senti] # which target
    rowi += 1
```

```python
lookuptable.shape, all_acts.shape
```

# Compress and project to 2D

```python
# PCA with 20 components just to show variance explained
scree = PCA(n_components=20).fit(all_acts).explained_variance_ratio_

# and plot
plt.figure(figsize=(8,3))
plt.plot(100*scree,'ks',markerfacecolor=[.9,.7,.7],markersize=10)

plt.gca().set(xlabel='Component number',xticks=range(0,21,2),
              ylabel='Percent variance explained',title='Scree plot')

plt.show()
```

```python
# re-run with only two components
# fit a PCA on all activations from all sentences and layers
pca = PCA(n_components=2).fit(all_acts)

# project to 2D
proj2d = pca.transform(all_acts)

print(all_acts.shape,proj2d.shape,lookuptable.shape)
```

# Average trajectories over sentences for each target word

```python
aveTrajectories = np.zeros((model.config.n_layer+1,2,3))

for i in range(model.config.n_layer+1):
  aveTrajectories[i,:,0] = proj2d[(lookuptable[:,1]==i) & (lookuptable[:,2]==0),:].mean(axis=0) # him
  aveTrajectories[i,:,1] = proj2d[(lookuptable[:,1]==i) & (lookuptable[:,2]==1),:].mean(axis=0) # her
  aveTrajectories[i,:,2] = proj2d[(lookuptable[:,1]==i) & (lookuptable[:,2]==2),:].mean(axis=0) # round


# distances between ave(him,her) and round
himher = aveTrajectories[:,:,:2].mean(axis=2)
round = aveTrajectories[:,:,-1]
gender_distances = np.sqrt(np.sum( (himher-round)**2 ,axis=1))

# calculate distances between "him" and "her" tokens
himher_distances = np.sqrt(np.sum(np.diff(aveTrajectories[:,:,:2],axis=2)**2,axis=1))
```

# Visualize!

```python
fig,axs = plt.subplots(1,2,figsize=(12,4))


# plot the trajectory for "round"
axs[0].plot(aveTrajectories[:,0,2],aveTrajectories[:,1,2],'g',zorder=-3,label='"round"')
axs[0].scatter(aveTrajectories[:,0,2],aveTrajectories[:,1,2],marker='s',
               s=np.linspace(20,120,nLayers),c=np.arange(nLayers),cmap='Greens')

# repeat for "him"
axs[0].plot(aveTrajectories[:,0,0],aveTrajectories[:,1,0],'b',zorder=-3,label='"him"')
axs[0].scatter(aveTrajectories[:,0,0],aveTrajectories[:,1,0],marker='o',
               s=np.linspace(20,120,nLayers),c=np.arange(nLayers),cmap='Blues')

# repeat for "her"
axs[0].plot(aveTrajectories[:,0,1],aveTrajectories[:,1,1],'r',zorder=-3,label='"her"')
h = axs[0].scatter(aveTrajectories[:,0,1],aveTrajectories[:,1,1],marker='^',
                   s=np.linspace(20,120,nLayers),c=np.arange(nLayers),cmap='Reds')

axs[0].set(xlabel='PC1',ylabel='PC2',title='State-space trajectories')
axs[0].legend()
fig.colorbar(h,ax=axs[0],label='Layer',pad=.01)



### show distances
axs[1].plot(gender_distances,color=[.7,.9,.7],linewidth=2,zorder=-6,label='"himher" vs. "round"')
axs[1].scatter(range(nLayers),gender_distances,marker='s',s=100,c=np.arange(nLayers),cmap='Greens')

axs[1].plot(himher_distances,color=[.9,.7,.7],linewidth=2,zorder=-3,label='"him" vs. "her"')
axs[1].scatter(range(nLayers),himher_distances,marker='o',s=100,c=np.arange(nLayers),cmap='RdPu')

axs[1].set(xlabel='Layer',ylabel='Euclidean distance',title='Inter-trajectory distances',
           ylim=[-1,None])
axs[1].legend()

plt.tight_layout()
plt.show()
```
