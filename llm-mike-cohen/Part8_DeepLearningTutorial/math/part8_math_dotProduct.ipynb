```python
import numpy as np
import torch
```

# Manual and numpy

```python
activations = np.array([ 1,3,-3,4,0 ])
weights = np.array([ .3,-.2,-.1,.4,.5 ])

# "linear weighted sum" = dot product
manual_dp = np.sum( activations * weights )
numpy_dp = np.dot(activations, weights)

print(manual_dp,numpy_dp)
```

# With pytorch

```python
activations = torch.tensor([ 1,3,-3,4,0 ])#,dtype=torch.float)
weights = torch.tensor([ .3,-.2,-.1,.4,.5 ])

manual_dp = torch.sum( activations * weights )
torch_dp = torch.dot(activations, weights)

print(manual_dp,torch_dp)
```
